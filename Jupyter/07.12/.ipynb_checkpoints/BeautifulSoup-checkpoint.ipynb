```python
#headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.106 Safari/537.36',}
#data=requests.get(url,headers=headers)

#파이썬으로 쿠팡에 접속하니깐, 막혀. 
# headers 는 나는 user 고 저런 정보로 접속하는 거야 라고 알려줘서 열리게 하는거야
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.106 Safari/537.36',}
coupang = requests.get("https://www.coupang.com/np/categories/194810" , headers = headers )
coupang.encoding = 'utf-8'
coupanghtml = coupang.text
coupangbs = bs4.BeautifulSoup(coupanghtml, 'html.parser')
```

### 테이블 크롤링

```python
#테이블크롤링은 패턴을 알고 접근을 해야해
import requests, bs4
import pandas as pd

#웹페이지 html 소스 가져오기
respspark = requests.get("https://sparkkorea.com/테스트/")
respspark.encoding='utf-8'
htmlspark = respspark.text
bsspark = bs4.BeautifulSoup(htmlspark, 'html.parser')
tabletag = bsspark.find("table", {"id":"test_table"})
```

```python
tabletag
```

```python
iter_tr = tabletag.find("tbody").find_all("tr")
iter_tr
```

```python
# 행별 컨텐츠를 담을 리스트
content=[]
# 각 행전체를 담을 리스트
allcontent=[]

for rows in iter_tr:
    tds = rows.find_all("td")
    for td in tds:
        content.append(td.text)
    allcontent.append(content)
    content=[]
allcontent
```

```python
pd.DataFrame(allcontent)
```
