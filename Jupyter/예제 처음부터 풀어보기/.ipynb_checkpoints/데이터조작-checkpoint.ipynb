```python
###실습 - 불량데이터 처리 
selloutData["QTY_NEW"] = np.where(selloutData["QTY"] < 0, 0, selloutData["QTY"])
```

```python
selloutData.head()
```

```python
###실습 -- 데이터 통합
#YEAR, WEEK 컬럼 생성하고, WEEK 가 52이하인 데이터만 refinedSelloutData 변수에 담아 
selloutData['YEAR'] = selloutData.YEARWEEK.str[0:4]
selloutData['WEEK'] = selloutData.YEARWEEK.str[4:]
```

```python
selloutData.dtypes
```

```python
refinedSelloutData = selloutData[selloutData.WEEK.astype(float)<= 52 ]
```

### 6.데이터 요약분석하기


```python
#데이터프레임명.groupby(['그룹컬렴명']).집계함수
#다중집계: 데이터프레임명.groupby(['그룹컬럼명']).agg(['mean','std'])["QTY_NEW"]
#특정컬럼에 대해서만 groupKey 적용시 , [[컬럼명]]
#집계함수: mean(), max(), min(), std()
#groupby 함수는 종료 후 reset_index() 를 통해 인덱스 초기화
#데이터프레임명.groupby(['컬럼명']).describe()
```

```python
#실습 -- EMI 별 MAX 값을 구하세요 
customerData.groupby(["EMI"]).max()
```

```python
###실습 -- refinedData 에서 지역, 상품, 연주차 단위로 정렬하여 sortedData 변수에 담아 
# groupKey = ['REGIONID','PRODUCT','YEARWEEK']
sortedData = refinedSelloutData.sort_values(['REGIONID','PRODUCT','YEARWEEK'])
```

```python
sortedData.head()
```

```python
###실습 -- [지역, 상품, 연도 별 집계] sortedData 에서 지역, 상품 단위 
         ##판매량(QTY_NEW)의 평균 연산 후 groupData 변수에 담으세요 
         ##이후 컬럼명을 QTY_MEAN 로 변경
```

```python
#groupKey = ['REGIONID','PRODUCT','YEAR']
```

```python
groupData = sortedData.groupby(['REGIONID','PRODUCT','YEAR'])[["QTY_NEW"]].mean()
```

```python
groupData["QTY_MEAN"] = groupData["QTY_NEW"]
groupData = groupData.drop("QTY_NEW",axis = 1) #
```
